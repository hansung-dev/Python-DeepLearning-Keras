## 3.1.6 plot 함수를 사용해 데이터 시각화하기

```python
import matplotlib.pyplot as plt

job_salary = df[["job_title", "salary"]]

job_salary.plot()
```

```python
job_usd_salary = df[["job_title", "salary_in_usd"]]

job_usd_salary.plot()
```

```python
job_salary_plot = job_usd_salary[job_usd_salary["job_title"].isin(["Data Scientist", "Data Engineer", "Data Analyst"])]

job_salary_plot.plot.scatter(x="job_title", y="salary_in_usd")
```

```python
job_usd_salary["salary_in_usd"].plot.box()
```

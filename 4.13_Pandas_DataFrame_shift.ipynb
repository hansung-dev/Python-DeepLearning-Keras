```python
import pandas as pd
import numpy as np

index = pd.date_range(start="2010-01-01", end="2021-12-31")
data = np.random.randint(0, 100, size=(len(index), 5))

df = pd.DataFrame(data=data, index=index, columns=["Open", "High", "Low", "Close", "Volume"])
df.head()
```

## 4.3.4 pandas.DataFrame.shift

```python
df.shift(periods=1).head()
```

```python
df.shift(periods=-1).tail()
```

```python
df.shift(periods=-1).head()
```

```python
df.shift(periods=2, fill_value=0).head()
```

```python
df.shift(periods=2, freq="D").head()
```

```python
df.tshift(periods=1, freq="D").head()
```

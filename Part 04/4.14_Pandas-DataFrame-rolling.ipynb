## 4-3-5 pandas.DataFrame.rolling

```python
import pandas as pd
import numpy as np

index = pd.date_range(start="2022-01-01", end="2022-01-05")
data = np.array(range(5))

df = pd.DataFrame(data=data, index=index, columns=["Close"])
df.head()
```

```python
df.rolling(window=2).sum()
```

```python
df.rolling(window=3).sum()
```

```python
df.rolling(window=2, min_periods=1).sum()
```

```python
df.expanding().sum()
```

```python
df['Close'].ewm(span=2).mean()
```

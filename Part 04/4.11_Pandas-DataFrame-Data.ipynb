## 4.3.2 주가 데이터셋 만들기

```python
import pandas as pd

index = pd.date_range("2022-01-01", periods=31, freq="D")

index
```

```python
import numpy as np

data = np.random.randint(0, 100, size=31)

df = pd.DataFrame(data=data, index=index, columns=["Close"]) 
print(df.head())
```

```python
df.plot()
```

```python
import numpy as np

data = np.random.randint(0, 100, size=(31, 5))

df = pd.DataFrame(data=data, index=index, columns=["Open", "High", "Low", "Close", "Volume"])
df.head()
```

```python
df[["Open", "Close"]].plot(figsize=(15, 5))
```

## 4.3.1 pandas.DatetimeIndex

```python
import pandas as pd 

pd.date_range(start="2022-01-01", end="2022-01-31")
```

```python
pd.date_range("2022-01-01", periods=31, freq="D")
```

```python
pd.date_range(start="2022-01-01", periods=3, freq="Y")
```

```python
pd.date_range("2022-01-01", periods=12, freq="M") 
```

```python
pd.date_range("2022-01-01", periods=24, freq="h")
```

```python
pd.date_range("2022-01-01", periods=60, freq="T")
```

```python
pd.date_range("2022-01-01", periods=60, freq="s")
```

```python
day = pd.date_range("2022-01-01", "2022-01-09", freq="D").to_series() 

day.dt.dayofweek
```

```python
day = pd.Series(pd.date_range("2022-02-27", periods=4)) 

print(day, '\n')

print(day.dt.is_month_start, '\n') 
print(day.dt.is_month_end)
```

## 4.3.3 pandas.DataFrame.resample

```python
import pandas as pd
import numpy as np

index = pd.date_range(start="2010-01-01", end="2021-12-31")
data = np.random.randint(0, 100, size=(len(index), 5))

df = pd.DataFrame(data=data, index=index, columns=["Open", "High", "Low", "Close", "Volume"])
df.head()
```

```python
df.groupby([df.index.year]).mean()
```

```python
df.groupby([df.index.year, df.index.month]).mean()
```

```python
# ruleDateOffset, Timedelta or str
# A : year end frequency

df.resample(rule='A').mean().head()
```

```python
# ruleDateOffset, Timedelta or str
# M : month end frequency

df.resample(rule='M').mean().head()
```

```python
df.resample(rule='M').min().head()
```

```python
df.resample(rule='M').max().head()
```

```python
df.resample(rule='7D').mean()
```

```python
df.resample('12H').mean()
```

```python
df.resample('12H').pad().head(6)
```

```python
df.resample('12H').bfill().head(6)
```

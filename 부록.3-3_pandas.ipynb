## 3.1.7 기존 열에서 새로운 열 생성하기

```python
job_salary = df[["Unnamed: 0", "work_year", "job_title", "salary_currency", "salary", "salary_in_usd"]].copy()
job_salary = job_salary[job_salary["salary_currency"] == "USD"]

# salary_in_eur = salary_in_usd * 0.97
job_salary["salary_in_eur"] = job_salary["salary_in_usd"] * 0.97

job_salary.head()
```

```python
job_salary = job_salary.rename(
    columns={
        "Unnamed: 0": "no",
    }
)

job_salary.head()
```

```python
job_salary.rename(columns=str.upper)
```

## 3.1.9 groupy를 사용해 데이터 그룹별 집계하기

```python
job_salary = df[["work_year", "job_title",  "salary_currency",  "salary", "salary_in_usd"]].copy()
job_salary["salary_in_eur"] = job_salary["salary_in_usd"] * 0.97

job_salary.head()
```

```python
job_salary[["work_year", "salary_in_usd", "salary_in_eur"]].groupby("work_year").mean()
```

```python
job_salary.groupby(["work_year", "salary_currency"]).mean()
```

```python
job_salary.groupby("salary_currency")["salary"].mean()
```

```python
job_salary.groupby(["job_title", "salary_currency"])["salary"].mean()
```

## 3.1.10 value_counts로 데이터 합계 확인하기

```python
job_salary.groupby("job_title")["job_title"].count().head()
```

```python
a = np.array([5, 10, 15])
b = np.array([5])

c = a - b
print(c)

d = c < 5
print(d)
```

```python
a = np.array([[1, 2],
              [3, 4]])
b = np.array([[1, 2],
              [3, 4]])
# 아다마르곱
print('elementwise product       \n', a * b)
# 행렬곱셈
print('matrix product            \n', a @ b)
# 스칼라곱
print('dot product               \n', a.dot(b))
```

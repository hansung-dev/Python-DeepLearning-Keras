## 1.1.6 Weight and Standing Height from NHANES

```python
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/NHANES Weight and Height/data.csv')

data.head()
```

```python
data.info()
```

```python
feature_cols = ['Weight (kg)', 'Standing Height (cm)']

label = data.pop('BMI(kg/m**2)')
feature = data[feature_cols].copy()

print(type(feature))
print(type(label))

feature.head()
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(feature, label, test_size=0.2)

print('train dataset :', train_x.shape, train_y.shape)
print('test dataset :', test_x.shape, test_y.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(units=2, activation='relu', input_shape=[2]),
    Dense(units=1, activation='relu'),
    Dense(units=1) # activation='linear'
])

model.compile(loss='mae',
              optimizer='adam',
              metrics=['mae', 'mse'])
```

```python
%%time 

model.fit(train_x, train_y, 
          epochs=1000)
```

```python
model.evaluate(test_x, test_y)
```

```python
x_test = [(91.1, 152.7)]
#y_test = [39.1]

print(model.predict(x_test))
```

##3.3.2 LSTM Layers

```python
import numpy as np

input_array = np.random.random([2, 3, 4]).astype(np.float32)

print("input_array.shape : ", input_array.shape)
print(input_array)
```

```python
from tensorflow.keras.layers import LSTM

lstm = LSTM(units=5)
output = lstm(inputs=input_array)

print("output.shape : ", output.shape)
print(output)
```

```python
from tensorflow.keras.layers import LSTM

lstm = LSTM(4, return_sequences=True, return_state=True)
timesteps_sequence_output, final_memory_state, final_carry_state = lstm(input_array)

print(timesteps_sequence_output.shape)
print(final_memory_state.shape)
print(final_carry_state.shape)
```

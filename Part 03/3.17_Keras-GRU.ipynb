## 3.3.3 GRU Layers

```python
import numpy as np

input_array = np.random.random([2, 3, 4]).astype(np.float32)

print("input_array.shape : ", input_array.shape)
print(input_array)
```

```python
from tensorflow.keras.layers import GRU

gru = GRU(units=5)

output = gru(input_array)
print(output.shape)
```

```python
from tensorflow.keras.layers import GRU

gru = GRU(units=5, return_sequences=True, return_state=True)
timesteps_sequence_output, final_state = gru(input_array)

print(timesteps_sequence_output.shape)
print(final_state.shape)
```
